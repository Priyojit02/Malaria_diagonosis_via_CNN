```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Define your dataset directory
dataset_directory = 'kyd_ids'

# Define image parameters
height, width, channels = 128, 128, 3
batch_size = 16

# Create an ImageDataGenerator for data augmentation and normalization
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Create a TensorFlow dataset from the generator
Train_dataset = tf.keras.utils.image_dataset_from_directory(
    dataset_directory,
    image_size=(height, width),
    batch_size=batch_size,
    shuffle=True,
    seed=42,
    validation_split=0.2,
    subset="training"
)

# Now you can iterate over the dataset
for i, (images, labels) in enumerate(Train_dataset.take(16)):
    for j in range(batch_size):
        ax = plt.subplot(4, 4, i * batch_size + j + 1)
        plt.imshow(images[j].numpy())
        plt.title(f"Label: {labels[j].numpy()}")
        plt.axis('off')

plt.show()


```
